## Create directories

```python
base_dir="Mask detection/"
```

```python
import splitfolders
splitfolders.ratio(base_dir, output="Mask_detection_data", seed=1337, ratio=(.8, 0.1,0.1)) 
```

```python
train_dir="Mask_detection_data/train"
val_dir="Mask_detection_data/val"
test_dir="Mask_detection_data/test"
```

Classes for Mask detection

```python
mask_classes=['fully_covered','not_covered','not_face','partially_covered']
```

## Using Image Generator from tensorflow 

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen=ImageDataGenerator(rescale=1./255,
                                   rotation_range = 40,
                                   width_shift_range = 0.2,
                                   height_shift_range = 0.2,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
val_datagen=ImageDataGenerator(rescale=1./255)
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
train_gen=train_datagen.flow_from_directory(train_dir,
                                           target_size=(64,64),
                                           batch_size=10,
                                           class_mode='categorical')
val_gen=val_datagen.flow_from_directory(val_dir,
                                           target_size=(64,64),
                                           batch_size=5,
                                           class_mode='categorical')
test_gen=train_datagen.flow_from_directory(test_dir,
                                           target_size=(64,64),
                                           batch_size=5,
                                           class_mode='categorical')
```

## Creating the model

```python
model=tf.keras.Sequential()
model.add(tf.keras.layers.Conv2D(32,(3,3),activation='relu',input_shape=(64,64,3)))
model.add(tf.keras.layers.MaxPool2D(2,2))
model.add(tf.keras.layers.Conv2D(64,(3,3),activation='relu'))
model.add(tf.keras.layers.MaxPool2D(2,2))
model.add(tf.keras.layers.Conv2D(128,(3,3),activation='relu'))
model.add(tf.keras.layers.MaxPool2D(2,2))
# model.add(tf.keras.layers.Conv2D(256,(3,3),activation='relu'))
# model.add(tf.keras.layers.MaxPool2D(2,2))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(256,activation='relu'))
model.add(tf.keras.layers.Dense(512,activation='relu'))
model.add(tf.keras.layers.Dense(4,activation='softmax'))
```

```python
model.compile(loss=tf.keras.losses.categorical_crossentropy,
             optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
             metrics=['accuracy'])
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
callbacks=EarlyStopping(
    monitor='val_accuracy', 
    patience=5
)
```

```python
EPOCHS=100

history_v1=model.fit(train_gen,
                     steps_per_epoch=529,
                     epochs=EPOCHS,
                     validation_data=val_gen,
                     validation_steps=132,callbacks=[callbacks])
```

```python
model.evaluate(test_gen)
```

```python
train_acc=history_v1.history['accuracy']
val_acc=history_v1.history['val_accuracy']
train_loss=history_v1.history['loss']
val_loss=history_v1.history['val_loss']

epochs=range(len(train_acc))

plt.figure(figsize=(10,7))

plt.plot(epochs,val_acc,color='red',label='validation accuracy')
plt.plot(epochs,train_acc,color='blue',label='train accuracy')
plt.title('Accuracy',fontsize=18)
plt.legend()

plt.figure(figsize=(10,7))

plt.plot(epochs,val_loss,color='red',label='validation loss')
plt.plot(epochs,train_loss,color='blue',label='trainloss')
plt.title('Loss',fontsize=18)
plt.legend()


plt.show()
```

## Creating the model with transfer learning

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras import layers
local_weights='inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5'
```

```python
pre_trained_model=InceptionV3(input_shape=(75,75,3),
                             include_top=False,
                             weights=None)
pre_trained_model.load_weights(local_weights)
for layer in pre_trained_model.layers:
    layer.trainable=False
```

```python
pre_trained_model.summary()
```

```python
last_layer = pre_trained_model.get_layer('mixed2')
print('last layer output shape: ', last_layer.output_shape)
last_output = last_layer.output
```

```python
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras import Model

x = layers.Flatten()(last_output)
x = layers.Dense(1024, activation='relu')(x)
x = layers.Dropout(0.2)(x)                  
x = layers.Dense(4, activation='softmax')(x)           

model_v1 = Model(pre_trained_model.input, x) 

model_v1.summary()
```

```python
model_v1.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001), 
              loss = 'categorical_crossentropy', 
              metrics = ['accuracy'])
```

```python
train_gen=train_datagen.flow_from_directory(train_dir,
                                           target_size=(75,75),
                                           batch_size=10,
                                           class_mode='categorical')
val_gen=val_datagen.flow_from_directory(val_dir,
                                           target_size=(75,75),
                                           batch_size=5,
                                           class_mode='categorical')
test_gen=test_datagen.flow_from_directory(test_dir,
                                           target_size=(75,75),
                                           batch_size=5,
                                           class_mode='categorical')
```

```python
EPOCHS=100

history_v2=model_v1.fit(train_gen,
                     steps_per_epoch=529,
                     epochs=EPOCHS,
                     validation_data=val_gen,
                     validation_steps=132,callbacks=[callbacks])
```

```python
model_v1.evaluate(test_gen)
```

```python
train_acc=history_v2.history['accuracy']
val_acc=history_v2.history['val_accuracy']
train_loss=history_v2.history['loss']
val_loss=history_v2.history['val_loss']

epochs=range(len(train_acc))

plt.figure(figsize=(10,7))

plt.plot(epochs,val_acc,color='red',label='validation accuracy')
plt.plot(epochs,train_acc,color='blue',label='train accuracy')
plt.title('Accuracy',fontsize=18)
plt.legend()

plt.figure(figsize=(10,7))

plt.plot(epochs,val_loss,color='red',label='validation loss')
plt.plot(epochs,train_loss,color='blue',label='trainloss')
plt.title('Loss',fontsize=18)
plt.legend()


plt.show()
```
